# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
data_set = pd.read_csv("placement_dataset.csv")
X = data_set.iloc[:,:-1].values #taking all rows of first col
y = data_set.iloc[:,-1].values #all rows of second col
```

# Spliting into Training and Testing Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

# Training on train set using Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Prediction on Test Set

```python
y_pred = regressor.predict(X_test)
```

# Visualizing the Training Set

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("CGPA vs Placement Exam Marks")
plt.xlabel("Pleacement Exam Marks")
plt.ylabel("CGPA")
plt.show()
```

# Visualizing the Testing Set

```python
plt.scatter(X_test, y_test, color = 'grey')
plt.plot(X_test, regressor.predict(X_test), color = 'green')
plt.title("CGPA vs Placement Exam Marks")
plt.xlabel("Pleacement Exam Marks")
plt.ylabel("CGPA")
plt.show()
```
