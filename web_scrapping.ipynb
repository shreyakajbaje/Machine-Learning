```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import Series, DataFrame
```

```python
url = 'https://www.studentnewsdaily.com/archive/daily-news-article/'
```

```python
#step 1: call api server using request module
result = requests.get(url)
soup = BeautifulSoup(result.content, "lxml")
#print(result.content)
```

```python
print(soup)
```

```python
#content = soup.findAll('div')
#print(content[0])

content = soup.findAll('section')
summary = content[1].find(class_='col-md-10 col-lg-10 col-xl-10')
print(summary)
```

```python
finaldata = summary.findAll('h4')
```

```python
series = [ele.get_text() for ele in finaldata]
```

```python
print(series)
```
